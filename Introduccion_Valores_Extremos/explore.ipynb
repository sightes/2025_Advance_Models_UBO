```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

```

Grafico dispersion 


CASO 1 : Alturas 

```python
url= os.path.join(os.getcwd(),  "data", "base_altura.xlsx")
data = pd.read_excel(url, sheet_name="Sheet 1")
```

```python
import plotly.express as px
fig = px.scatter(x=data.ID, y=data.Altura)
mediana = data['Altura'].median()
fig.add_shape(
    type='line',
    x0=data['ID'].min(), x1=data['ID'].max(), 
    y0=mediana, y1=mediana,  
    line=dict(color='red', width=2, dash='dash'),  
    name='Mediana'
)

fig.update_layout(
    width=1000,  # Ancho en píxeles (equivalente a 10 en Matplotlib)
    height=500,  # Alto en píxeles (equivalente a 5 en Matplotlib)
    title_x=0.5  # Centrar el título
)
fig.show()
```

```python
import plotly.express as px
fig = px.histogram(data, x="Altura",nbins=30)
fig.update_layout(bargap=0.2)
fig.show()
```
