```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Generación de datos sintéticos
np.random.seed(42)
data = np.concatenate([
    np.random.normal(50, 10, 1000),  # Datos normales
    np.random.normal(150, 5, 10)    # Outliers
])

# 2. Selección del umbral para Generalized Pareto Distribution (GPD)
# Seleccionamos el percentil 90 como umbral inicial
threshold = np.percentile(data, 90)
excess_data = data[data > threshold] - threshold

# 3. Ajuste de la GPD
shape_gpd, loc_gpd, scale_gpd = stats.genpareto.fit(excess_data)

# Interpretación del parámetro de forma (xi) para seleccionar la distribución
if shape_gpd > 0:
    selected_distribution = "Fréchet"
elif np.isclose(shape_gpd, 0, atol=1e-2):  # Aproximación a cero
    selected_distribution = "Gumbel"
else:
    selected_distribution = "Weibull"

print("Parámetro de forma (xi):", shape_gpd)
print(f"Distribución seleccionada según GPD: {selected_distribution}")

# 4. Ajuste de la distribución seleccionada
x = np.linspace(min(data), max(data), 1000)

if selected_distribution == "Fréchet":
    shape, loc, scale = stats.genextreme.fit(data, -0.5)  # xi negativo para Fréchet
    pdf = stats.genextreme.pdf(x, -shape, loc, scale)
elif selected_distribution == "Gumbel":
    loc, scale = stats.gumbel_r.fit(data)
    pdf = stats.gumbel_r.pdf(x, loc, scale)
else:  # Weibull
    shape, loc, scale = stats.weibull_min.fit(data)
    pdf = stats.weibull_min.pdf(x, shape, loc, scale)

# 5. Visualización de los datos y la distribución seleccionada
plt.hist(data, bins=30, density=True, alpha=0.5, label="Datos")
plt.plot(x, pdf, label=f"Distribución seleccionada: {selected_distribution}", lw=2)
plt.legend()
plt.title("Selección de distribución basada en GPD")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.show()

# 6. Detección de outliers con la distribución seleccionada
if selected_distribution == "Fréchet":
    p95 = stats.genextreme.ppf(0.95, -shape, loc, scale)
elif selected_distribution == "Gumbel":
    p95 = stats.gumbel_r.ppf(0.95, loc, scale)
else:  # Weibull
    p95 = stats.weibull_min.ppf(0.95, shape, loc, scale)

outliers = data[data > p95]
print("Percentil 95:", p95)
print("Outliers detectados:", outliers)

# 7. Visualización de los outliers
plt.boxplot([data, outliers], labels=["Datos", "Outliers"])
plt.title("Outliers detectados según la distribución seleccionada")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generación de datos sintéticos (distribución de Pareto)
np.random.seed(42)
data = np.random.pareto(a=3, size=1000) + 1  # Distribución de Pareto con forma α=3

# 2. Ordenar los datos de mayor a menor
sorted_data = np.sort(data)[::-1]

# 3. Definir el número de valores extremos (k)
k_values = np.arange(10, 200)  # Rango de k para calcular Hill
hill_estimates = []

# 4. Cálculo del estadístico de Hill para diferentes k
for k in k_values:
    top_k = sorted_data[:k]
    threshold = sorted_data[k]
    hill_stat = (1 / k) * np.sum(np.log(top_k / threshold))
    hill_estimates.append(hill_stat)

# 5. Visualización de Hill vs k
plt.plot(k_values, hill_estimates, label="Estimación de Hill (\u03BE)")
plt.axhline(y=1/3, color='r', linestyle='--', label="Valor teórico (\u03BE = 1/3)")
plt.xlabel("Número de valores extremos (k)")
plt.ylabel("Estimación de Hill (\u03BE)")
plt.title("Estadístico de Hill para diferentes k")
plt.legend()
plt.show()

# 6. Resultado para un valor de k específico
k_opt = 50  # Por ejemplo, k = 50
hill_xi = hill_estimates[k_opt - k_values[0]]
print(f"Estimación de Hill para k={k_opt}: \u03BE ≈ {hill_xi:.4f}")
```
